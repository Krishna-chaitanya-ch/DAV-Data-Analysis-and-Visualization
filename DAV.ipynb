```python
# Importing the library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# reading the Excel file data  & Display contents
df = pd.read_excel('/content/Bank data.xlsx')
df
```

```python
# Retrieve the dimensions of the DataFrame (rows, columns)
df.shape
```

```python
# Display a concise summary of the DataFrame, including column names, data types, non-null counts, and memory usage.
df.info()
```

```python
# Convert the 'column_name' from object to float
df['EstimatedSalary'] = df['EstimatedSalary'].astype(float)
df['Balance']=df['Balance'].astype(float)
df.info()
```

```python
df.head()
```

```python
pd.isnull(df).sum()
```

# DATA CLEANING

```python
df['Balance'].fillna(df['Balance'].mean(), inplace=True)
```

```python
df['EstimatedSalary'].fillna(df['EstimatedSalary'].mean(), inplace=True)
```

```python
df.head()
```

```python
df[df['Loan Taken']=='no']
```

```python
df.head()
```

```python
pd.isnull(df).sum()
```

```python
df.fillna('nan', inplace=True)
```

```python
df.head()
```

# 1 heatmap
Computeing and visualize the correlation matrix for the numerical columns in the DataFrame using a heatmap.


```python
numeric_columns = df.select_dtypes(include=['number'])
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

# 2 regplot
Developing a regression model to analyze and visualize the relationship between CustomerId & Age.

```python
sns.regplot(x='EstimatedSalary', y='Age', data=df)
plt.show()
```

# 3 to_datetime
Converting the index to datetime format for improved time-based analysis.

```python
df.index = pd.to_datetime(df.index)
df.index
```

# 4 displot
Creating a distribution plot (displot) to visualize the distribution of Loan Taken values.

```python
sns.displot(df['Loan Taken'], kde=True)
```

# 5 Histogram
Creating a histogram to visualize the distribution of Age values.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title('Age Distribution')
```

# 6 Countplot
Creating a count plot to visualize the distribution of Loan Cleared values.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Loan cleared', data=df)
plt.show()
```

# 7 Bar chart
Creating a bar plot to visualize the relationship between the Exited and Loan Cleared variables.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Exited', y='Loan cleared', data=df)
plt.show()
```

# 8 Lineplot
Creating a line plot to visualize the relationship between Age and 'CreditScore.

```python
sns.lineplot(x='Age', y='CreditScore', data=df)
```

# 9 violinplot
Creating a violin plot to visualize the distribution of the Balance variable.


```python
sns.violinplot(data=df['EstimatedSalary'], orient='vertical', color='orange')
```

# 10 scatterplot
Creating a scatter plot to visualize the relationship between Age and Location.

```python
sns.scatterplot(x='Age', y='Location', data=df, color='Blue')
```

# 11 pie chart
Creating a pie chart to visualize the distribution of Age values.

```python
plt.figure(figsize=(8, 6))
df['Age'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.show()
```

# 12 Sorting the DataFrame by the Location column in ascending order and displaying the first 10 rows.

```python
df.sort_values(by=['Location'], ascending=True).head(10)
```

# 13 pie chart
Creating a pie chart to visualize the distribution of Exited values.


```python
plt.figure(figsize=(8, 6))
df['Exited'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Exited')
plt.show()
```

# 14 value_counts
Counting the occurrences of each unique CustomerName.


```python
df["Location"].value_counts()
```

# 15 FacetGrid and scatterplot
Creating a facet grid of scatter plots for each unique Location to visualize the relationship between Age & CreditScore.

```python
grid=sns.FacetGrid(df, row='Location')
grid.map(sns.scatterplot,'Age','CreditScore',color='blue')
```

# 16 heatmap
Creating a heatmap to visualize the CreditScore column with annotated values.

```python
sns.heatmap(df[['CreditScore']])
```

# 17 countplot
Creating a count plot to visualize the frequency distribution of Gender values.

```python
sns.countplot(x='Exited',hue='Gender', data=df)
```

# 18 FacetGrid and scatterplot
Creating a facet grid of scatter plots to visualize the relationship between Age & CreditScore for each Exited value.

```python
grid=sns.FacetGrid(df, col='Exited')
grid.map(sns.scatterplot,'Age','CreditScore')
```

# 19 Box plot
Creating a box plot to visualize the distribution and spread of values across all numerical columns.

```python
df.plot(kind='box', figsize=(10, 6))
plt.ylabel('Values')
plt.show()
```

# 20 Correlation between the features
Computing and visualizing the correlation matrix for numerical columns in the DataFrame using a heatmap.

```python
numeric_columns = df.select_dtypes(include=['number'])
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

Creating a violin plot to visualize the distribution and density of CreditScore values.
